```python
def add_control_features(df):
    df = df.copy()

    # Previous error
    df['Error_prev'] = df['Error'].shift(1).fillna(0.0)

    # Leaky integral
    df['E_int'] = 0.0
    for i in range(1, len(df)):
        df.loc[df.index[i], 'E_int'] = (
            df.loc[df.index[i-1], 'E_int']
            + df.loc[df.index[i], 'Error']*0.01
        )


    # Previous PWM
    df['PWM_prev'] = df['PWM'].shift(1).fillna(0.0)
    df['delta_PWM'] = df['PWM'] - df['PWM_prev']

    return df
dfs = [df_first, df_second, df_fourth, df_fifth, df_sixth]

dfs = [add_control_features(df) for df in dfs]
```

```python
import pandas as pd

def split_df_80_10_10(df):
    n = len(df)

    n_train = int(0.8 * n)
    n_val   = int(0.1 * n)

    df_train = df.iloc[:n_train].copy()
    df_val   = df.iloc[n_train:n_train + n_val].copy()
    df_test  = df.iloc[n_train + n_val:].copy()

    return df_train, df_val, df_test
```

```python
dfs_new = {
    "sine": dfs[1],
    "step_up": dfs[2],
    "step_down": dfs[3],
    "zero_sp_zero": dfs[4],
    "random": dfs[0],
}

train_list, val_list, test_list = [], [], []

for name, df in dfs_new.items():
    if name == "random":
        # Keep all random data in training
        train_list.append(df)
    else:
        tr, va, te = split_df_80_10_10(df)
        train_list.append(tr)
        val_list.append(va)
        test_list.append(te)

df_train = pd.concat(train_list, ignore_index=True)
df_val   = pd.concat(val_list, ignore_index=True)
df_test  = pd.concat(test_list, ignore_index=True)

print (df_train)
print (df_val)
print (df_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import Huber
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.metrics import MeanSquaredError,MeanAbsoluteError, RootMeanSquaredError

def create_model():
    model = Sequential([
        Dense(16, activation='relu', input_shape=(6,)),
        Dense(16, activation='relu'),
        Dense(16, activation='relu'),
        Dense(1, activation='linear')   # PWM output
    ])
    model.compile(
        optimizer=Adam(learning_rate=1e-3),
        loss='mse',
        metrics=[MeanAbsoluteError(name='mae'), RootMeanSquaredError(name='rmse')]
    )
    model.summary()
    return model
```

```python
from tensorflow.keras.models import load_model

model = load_model("/content/drive/MyDrive/run12_1e-3.keras")

import json

weights_dict = {}

for i, layer in enumerate(model.layers):
    w = layer.get_weights()
    if w:
        weights_dict[layer.name] = {
            "W": w[0].tolist(),
            "b": w[1].tolist()
        }

with open("/content/drive/MyDrive/run12_1e-3.json", "w") as f:
    json.dump(weights_dict, f, indent=2)

```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',      # what to watch
    patience=10,             # epochs with no improvement
    min_delta=0,          # minimum improvement
    restore_best_weights=True
)
```

```python
for run in range(20):
    print(f"=== Run {run+1}/20 ===")
    model = create_model()  # Fresh model each run
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=500,
        batch_size=256,
        shuffle=False,  # Important for control data
        callbacks=[early_stop]
    )
    model.save(f"/content/drive/MyDrive/run{run}_1e-3.keras")
```
